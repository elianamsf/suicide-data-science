```python
import pandas as pd
```

```python
# Quando forem utilizar essas funções, especificamente prepare_df e colunms_transp,
# estudem o csv primeiro.
# É importante conhecer a estrutura dos arquivos que vamos trabalhar.

def prepare_df(csv = '', skip_rows = None, drop_columns = [], axis = 1 ):    
    data_frame = pd.read_csv(csv, skiprows = skip_rows)
    data_frame = data_frame.drop(labels = drop_columns, axis = axis)   
    return data_frame

def columns_transp (frame,  column_base='', convert_columns =[], var_name=None, value_name='value'):
    data_frame = pd.melt(frame, id_vars=[column_base], value_vars=convert_columns, var_name= var_name, value_name=value_name)
    return data_frame
    
```

```python
# Ler o csv que não esteja codificado em utf-8
def read_base(csv, skiprows = None):
    df = pd.read_csv(csv, encoding = 'iso-8859-1', skiprows = skiprows )
    return df
```

```python
# Exporta o csv sem criar colula pra o index:
def export_csv(data_frame, csv_address = None):
    data_frame.to_csv(csv_address, index = False)
```

```python
# As células a baixo são apenas um tutorial de execução:
# Caminho para encontrar o csv desse exemplo: CVS's->Convertidos
```

```python
# Lendo o database e excluindo a primeira linha do csv:
# Por que excluir a primeira linha?
# O pandas indentifica na primeira linha do csv o nome das colunas(atributos);
# Quando a coluna não possui nome, o pandas seta o valor como "Unnamed: X".
# Onde X é o número da coluna (começando por 0).
# Excluirmos a primeira linha, no caso desse csv, utilizado para testes, pois o nome das colunas estão na segunda linha.
# Então sempre verifique em qual linha está o nome de suas colunas(atributos), para eliminar linhas não necessárias.
# E como no nosso exemplo, eliminar as colunas vazias, como veremos em seguida.
# * Parâmetro utilizado para excluir linhas: skiprows.

read_base('../csvs/Human Development Index (HDI), female.csv',1).head()

#Função head(): Lendo apenas as 5 primeiras linhas. No nosso caso, apenas para vizualização. Faremos assim daqui em diante. ;)
```

```python
# Depois de ler o arquivo verificamos que algumas colunas estão vazias e são marcadas como
# "Unnamed: X".
#  Onde X é o número da coluna (começando por 0).
#  Chamamos a função prepare_df que lê o arquivo e elimina as colunas em branco.
#  Parâmetros: 
#   csv: Endereço absoluto ou relativo da localização do seu csv no seu sistema de arquivos;
#   skiprows: Explicado acima;
#   drop_colunms: lista de colunas que serão excluidas (Estas quais, foram indeticadas acima);
#   axis: 0 para index e 1 para colunas.
#  Retorno: Um DataFrame
prepare_df('../csvs/Human Development Index (HDI), female.csv', 1, 
                        ["Unnamed: 3","Unnamed: 5", "Unnamed: 7",
                          "Unnamed: 9", "Unnamed: 11", "Unnamed: 13",
                         "Unnamed: 15", "Unnamed: 17","Unnamed: 19",
                         "Unnamed: 21", "Unnamed: 23"], ).head()
```

```python
# Nada novo aqui, estamos apenas criando uma variavel do tipo DataFrame,
# para utilizarmos na função que inverte as colunas do csv, utilizando a função explicada acima.
data_base = prepare_df('../csvs/Human Development Index (HDI), female.csv', 1, 
                        ["Unnamed: 3","Unnamed: 5", "Unnamed: 7",
                          "Unnamed: 9", "Unnamed: 11", "Unnamed: 13",
                         "Unnamed: 15", "Unnamed: 17","Unnamed: 19",
                         "Unnamed: 21", "Unnamed: 23"], )
data_base.head()
```

```python
# Parâmetro 1 (frame): Passamos o dataframe que queremos tratar;
# Parâmetro 2 (colunm_base): A coluna que irá servir de base para as demais colunas;
# Parâmetro 3 (convert_columns): Uma lista de colunas que queremos transformar em uma coluna somente, para cada item da coluna base;
# Parâmetro 4 (var_name): O nome da coluna irá receber as colunas que virarão valores;
# Parâmetro 5 (value_name): O nome da coluna que irá receber os valores desejados.
db_ivertido = columns_transp(data_base,'Country',['2000','2005','2010','2011',
                                    '2012','2013','2014','2015'],
                                    "Ano", "HDI")
db_ivertido.head()
```

```python
# Exporta o csv:
export_csv(db_ivertido, '../csvs/HDIfemale.csv')
```
